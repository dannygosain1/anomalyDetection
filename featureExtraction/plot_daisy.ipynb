```python
import numpy as np
import sklearn
import matplotlib.pyplot as plt
from skimage.feature import daisy
from skimage import data
from skimage.io import imread
from PIL import Image
from sklearn.decomposition import PCA

file_text = "" # the final libsvm format data is stored here
combined_descs = [] # list of lists for the descs of each frame

for num in range(0, 42): #hardcoded values for now. Add 88 to account for deleted frames in the start

    image_location = "../imageExtraction/eagle/output-" + str(num) + ".jpg" # location of the image extraction
    print(image_location)

    print("creating descs")
    image = Image.open(image_location)
    convertedImg = image.convert('L') # converting rgb to grayscale
    img = convertedImg
    
    # creating descs for the each image using dense daisy feature extraction method
    descs, descs_img = daisy(img, step=45, radius=30, rings=2, histograms=6,
                         orientations=8, visualize=True)
    
    # adding the descs to the final list
    for x in range(0,descs.shape[0]):
        for y in range(0,descs.shape[1]):
            combined_descs.append(descs[x,y,:])

# pca svd
X = np.array(combined_descs)
pca = PCA(n_components=100) # number of descriptors, the component is an arbitrary number
# pca = PCA(n_components=descs.shape[2], svd_solver='full')
pca._fit(X)

# converting to libsvm format
print("converting to libsvm format")
file_text += str(num+1) + " "
for index,value in enumerate(pca.explained_variance_ratio_):
    file_text += str(index+1) + ":" + str(value) + " "
file_text += "\n"

# to view the dense daisy historgrams and rings on the frames

# fig, ax = plt.subplots()
# ax.axis('off')
# ax.imshow(descs_img)
# descs_num = descs.shape[0] * descs.shape[1]
# ax.set_title('%i DAISY descriptors extracted:' % descs_num)
# plt.show()
```
